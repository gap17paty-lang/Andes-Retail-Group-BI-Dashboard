### Narrativa del Dashboard (Modelo SCQA)
Se describe de forma estructurada la historia completa del dashboard usando SCQA, incluyendo contexto, problemas detectados, preguntas de negocio y recomendaciones basadas en datos.

#### 🖥️ Vista General (Overview)

**S (Situación):**  
El negocio cerró el periodo con un ingreso total consolidado de $5,531,994. El análisis temporal muestra un desempeño robusto durante los extremos del año (enero y diciembre), con ingresos que superan los $770k mensuales. Asimismo, las categorías de Deportes y Electrónica se mantienen como los pilares de facturación, aportando el mayor volumen al mix de ventas.

**C (Complicación):**  
A pesar del volumen global, se identifican dos señales de alerta críticas:

Debilidad Estacional: Existe una caída drástica en el rendimiento durante el segundo trimestre, alcanzando un "valle" mínimo en julio ($212,222), lo que representa casi una cuarta parte de lo recaudado en enero.

Deserción de Valor: El segmento Premium sufrió una contracción de $196,131 entre 2024 y 2025, siendo el único segmento con una caída de doble dígito, mientras que el resto de las categorías de producto (Deportes, Ropa, etc.) muestran una tendencia a la baja o estancamiento en el último año.

**Q (Pregunta):**   

¿Cómo podemos mitigar la volatilidad estacional del periodo junio-agosto y qué acciones de fidelización debemos ejecutar para recuperar la participación del segmento Premium, que es vital para el margen de rentabilidad?

**A (Respuesta):**  
La estrategia debe dividirse en dos ejes:

Estabilización Temporal: Lanzar campañas promocionales específicas o diversificación de inventario para los meses de verano (Q2 y Q3) para elevar el suelo de ventas.

Plan de Rescate Premium: Implementar un programa de beneficios exclusivos y revisión de precios en las categorías de Deportes y Electrónica (las de mayor impacto) para incentivar el retorno de los clientes de alto valor y frenar la migración hacia competidores o segmentos inferiores.

#### 🔎 Vista Detalle

**S (Situación):**  
A nivel regional, el mercado está liderado por Perú y Chile, con ingresos que superan los $2.1M y $2.0M respectivamente. El análisis de jerarquía muestra que el grueso de la rentabilidad ("Venta Alta") proviene de los segmentos Premium y Estándar en todas las categorías, destacando Deportes y Electrónica como los productos con mayor tracción en el segmento de alto valor.

**C (Complicación):**  
Se identifica una brecha considerable en el mercado de Colombia, cuya facturación ($1.3M) es significativamente menor a la de sus pares. Además, al observar la "Venta Baja", se detecta una acumulación preocupante de bajos ingresos en el segmento Económico para todas las categorías (específicamente en Hogar y Deporte), lo que sugiere que el volumen masivo de clientes no está compensando la rentabilidad necesaria.

**Q (Pregunta):**  
¿Cómo podemos elevar el desempeño comercial en Colombia para alinearlo con Perú y Chile, y qué ajustes de catálogo o precio son necesarios en el segmento Económico para convertir esas "Ventas Bajas" en oportunidades de mayor margen?

**A (Respuesta):**  
Se recomienda una auditoría comercial en la región de Colombia para identificar barreras de entrada o falta de penetración. Paralelamente, se debe optimizar la estrategia del segmento Económico, posiblemente mediante cross-selling (ventas cruzadas) con productos de las categorías líderes (Deportes o Electrónica) para elevar el ticket promedio en los grupos de venta baja.

**📊 Actualización desempeño comercial**

Equipo, les comparto los hallazgos significativos tras el análisis profundo del cierre del bienio en Tableau. Aquí los puntos críticos para nuestra siguiente sesión estratégica:

### 1. Insights Destacados (Lo que los datos nos dicen)
Fuga en el Segmento Premium: Es nuestra mayor alerta. Perdimos casi $200k en ingresos de clientes de alto valor entre 2024 y 2025.

Dominio Regional: Perú y Chile son nuestros motores principales ($2M+ c/u), mientras que Colombia presenta una brecha de oportunidad importante (está un 35% por debajo de Perú).

Estacionalidad valorativa: Tenemos un "valle" de ventas muy marcado entre junio y agosto. Enero y diciembre siguen siendo nuestros picos históricos.

Mix de Producto: Deportes y Electrónica lideran la rentabilidad, pero el segmento Económico está estancado en "Venta Baja" en todas las categorías.

### 2. Impacto en el Negocio (¿Por qué nos importa?)
Riesgo de Rentabilidad: Al caer el segmento Premium, el margen de beneficio se comprime. Necesitamos vender mucho más volumen "Económico" para compensar la pérdida de un solo cliente "Premium", lo cual no es sostenible a largo plazo.

Dependencia Geográfica: El negocio depende fuertemente de dos mercados. Cualquier cambio regulatorio o económico en Perú o Chile impactaría el 75% de nuestros ingresos.

### 3. Acciones Recomendadas (Roadmap)
Programa de Retención Premium: Se sugiere implementar una estrategia de beneficios exclusivos para recuperar a los clientes de alto valor en las categorías de Deportes y Electrónica.

🇨🇴 Plan de Expansión Colombia: Auditar la operación en Colombia para entender por qué no estamos llegando a los niveles de Perú/Chile.

Estrategia "Summer Boost": Diseñar campañas de marketing agresivas para Q2 y Q3 (junio-agosto) para estabilizar los ingresos durante los meses de baja estacionalidad.

Optimización del Ticket Promedio: En el segmento Económico, aplicar tácticas de cross-selling con productos de alta rotación para elevar el valor de cada transacción.
